```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None) 
pd.set_option("display.max_colwidth", None)
```

```python
playerinfo_df = pd.read_csv("../tests/test_data/test_cleaned_playerinfo.csv")
salaries_df = pd.read_csv("../tests/test_data/test_cleaned_salaries.csv")

playerinfo_df
# salaries_df
```

```python
# Merge both DataFrames into one
player_info_and_salaries_df = pd.merge(playerinfo_df, salaries_df, how="inner", on="player_name")

player_info_and_salaries_df = player_info_and_salaries_df[[
    "season_start_year",
    "player_name",
    "position",
    "height",
    "weight",
    "birth_date",
    "weight_kg",
    "height_m",
    "salary",
    "inflation_adjusted_salary"
]]

player_info_and_salaries_df.info()
```

```python
# Save player information to CSV
FILE_PATH = "../tests/test_data/test_player_info_and_salary.csv"
player_info_and_salaries_df.to_csv(FILE_PATH, index=False)
```
