```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None) 
pd.set_option("display.max_colwidth", None)
```

```python
boxscore_df = pd.read_csv("../tests/test_data/test_cleaned_boxscores.csv")
games_df = pd.read_csv("../tests/test_data/test_cleaned_games.csv")
```

```python
games_subset = games_df[["game_id", "season_start_year", "home_team", "points_home", "away_team", "points_away"]]
merged_df = boxscore_df.merge(games_subset, how="inner", on="game_id")

merged_df.head()
```

```python
# Who scored the most 3 pointers per year (regular season and pre season and friendlies)?
player_year_totals = merged_df.groupby(['season_start_year', 'player_name'])['three_pointers'].sum().reset_index()

top_scorers_all_years = player_year_totals.loc[player_year_totals.groupby('season_start_year')['three_pointers'].idxmax()]

top_scorers_all_years.head(5)
```
